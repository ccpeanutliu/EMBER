```python
import numpy as np
import keras
from keras.models import Sequential, load_model
from keras.layers import Conv1D, MaxPooling1D, LSTM, GRU, Dense, Dropout, Embedding
from keras.layers.normalization import BatchNormalization
from keras.layers.advanced_activations import LeakyReLU
from keras.callbacks import EarlyStopping
#from gensim.models import Word2Vec
```

```python
def model():
    np.random.seed(1230)
    model = Sequential()
    #model.add(embedding_layer)
    model.add(Conv1D(filters = 64, kernel_size = 8, strides = 2, activation = 'relu',padding='same', input_shape=(4,256)))
    model.add(Conv1D(filters = 128, kernel_size = 8, strides = 2, activation = 'relu',padding='same'))
    model.add(LeakyReLU(alpha=0.03))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size = 4, padding='same'))
    model.add(Dropout(0.5))
    model.add(Conv1D(filters = 128, kernel_size = 8, strides = 2, activation = 'relu',padding='same'))
    model.add(Conv1D(filters = 256, kernel_size = 8, strides = 2, activation = 'relu',padding='same'))
    model.add(LeakyReLU(alpha=0.03))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size = 4, padding='same'))
    model.add(Dropout(0.5))
    model.add(LSTM(64))
    model.add(Dense(256))
    model.add(Dropout(0.3))
    model.add(Dense(1, activation='sigmoid'))
    return model
```
