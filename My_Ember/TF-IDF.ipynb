```python
import numpy as np
import jsonlines
```

```python
num = 0
path = "./ember_dataset/ember/train_features_"+ str(num) +".jsonl"
```

```python
origin_feature = []
with jsonlines.open(path) as reader:
    for obj in reader:
        origin_feature.append(obj)
n = len(origin_feature)
```

```python
tf = []
idf = {}
count = np.zeros(n)

for i in range(n):
    tf.append({})

for i in range(n):
    b = True
    tmp = origin_feature[i]['imports'].keys()
    for obj in tmp:
        tmp_n = len(origin_feature[i]['imports'][obj])
        for j in range(tmp_n):
            tmp_str = origin_feature[i]['imports'][obj][j]
            count[i] += 1
            if tmp_str not in tf[i]:
                tf[i][tmp_str] = 1
                idf[tmp_str] = 1
            else:
                tf[i][tmp_str] += 1
                if b:
                    idf[tmp_str] += 1
                    b = False
for i in range(n):                    
    for obj in tf[i].keys():
        tf[i][obj] = tf[i][obj] / count[i]
for obj in idf.keys():
    idf[obj] = np.log10(n/idf[obj])
```

```python
print(count)
a = 0
tf_idf = []
for i in range(n):
    tf_idf.append({})

for i in range(n):
    for obj in tf[i].keys():
        a += 1
        tf_idf[i][obj] = tf[i][obj] * idf[obj]
        if a > 100 and a < 500:
            print(tf_idf[i][obj])
```

```python
m = 0
save = 0
for i in range(n):
    try:
        if m < tf_idf[i]['SetFileTime']:
            m = tf_idf[i]['SetFileTime']
            save = i
            print(save)
    except:
        continue
print(save)
```
