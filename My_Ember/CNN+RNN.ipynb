```python
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, LSTM, GRU, Dense, Dropout
from keras.layers.normalization import BatchNormalization
from keras.layers.advanced_activations import LeakyReLU
from keras.callbacks import EarlyStopping
```

```python
def model():
    np.random.seed(1230)
    model = Sequential()
    model.add(Conv1D(filters = 64, kernel_size = 8, strides = 1, activation = 'relu',padding='same', input_shape=(4,256)))
    model.add(Conv1D(filters = 64, kernel_size = 8, strides = 1, activation = 'relu',padding='same'))
    model.add(LeakyReLU(alpha=0.03))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size = 4, padding='same'))
    model.add(Dropout(0.3))
    model.add(Conv1D(filters = 128, kernel_size = 8, strides = 1, activation = 'relu',padding='same'))
    model.add(Conv1D(filters = 128, kernel_size = 8, strides = 1, activation = 'relu',padding='same'))
    model.add(LeakyReLU(alpha=0.03))
    model.add(BatchNormalization())
    model.add(MaxPooling1D(pool_size = 4, padding='same'))
    model.add(Dropout(0.3))
    model.add(GRU(64))
    model.add(Dense(64))
    model.add(Dense(128))
    model.add(Dropout(0.3))
    model.add(Dense(1, activation='sigmoid'))
    return model
```

```python
model = model()
for itr in range(6):
    
    print("Round",itr,"\n")
    
    data = np.load("./feature/all_features"+str(itr)+".npy")
    label = np.load("./label/"+str(itr)+".npy")
    delete = []
    n = label.shape[0]
    # delete byteentropy
    print("Deal data...\n")
    data = np.delete(data,2,axis = 1)
    
    for i in range(n):
        if label[i] == -1:
            delete.append(i)
    new_label = np.delete(label, delete)
    new_data = np.delete(data, delete, axis=0)
    print("Start training...\n")
    model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
    #es = EarlyStopping(monitor='val_loss', patience=4, verbose=0, mode='min')
    model.fit(x=new_data,y=new_label, batch_size=3000, epochs=50, verbose=1,validation_split=0.2, shuffle = True)#, callbacks=[es])
```
